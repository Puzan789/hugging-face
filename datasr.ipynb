
In Python, the yield statement is used in the context of a generator function. A generator function is a special type of function that allows you to iterate over a potentially large sequence of data without loading the entire sequence into memory at once. The yield statement is used to produce a series of values one at a time, and it temporarily suspends the function's state, allowing it to be resumed from that state later.

```python
squad=load_dataset("squad",split="train")
squad.keys()
```

```python
def get_training_corpus():
    datasets=squad["train"]
    for start_idx in range(0,len(datasets),1000):
        samples=datasets[start_idx:start_idx +1000]
        yield samples["w"]
```

```python
from transformers import AutoTokenizer
training_corpus=get_training_corpus()
old_tokenizer=AutoTokenizer.from_pretrained("gpt2")
new_tokenizer=old_tokenizer.train_new_from_iterator(training_corpus,52000)
```

# Why are fasttokenizers fast


```python
from datasets import load_dataset
raw_datasets=load_dataset("glue","mnli")
raw_datasets
```

```python
from transformers import AutoTokenizer
fast_tokenizer=AutoTokenizer.from_pretrained("bert-base-cased")
def tokenize_with_fast(examples):
    return fast_tokenizer(
    examples['premise'],examples['hypothesis'],truncation=True
    )
```

```python
%time tokenized_datasets=raw_datasets.map(tokenize_with_fast,batched=True)
```

```python
from transformers import pipeline
token_classifier=pipeline("token-classification")
token_classifier("MY name is Salysialix and I work at a laughing face in Brooklyn ",aggregation_strategy="simple")
```

```python
from transformers import pipeline

token_classifier = pipeline("token-classification")
result = token_classifier("My name is Don, and I work at a big laughing face company am a don")

print(result)
```

```python
qa=pipeline("question-answering")
context=""" 
K P Sharma Oli has faced allegations of corruption during his political career, with critics accusing him of engaging in unethical practices. These accusations have raised concerns about his leadership and governance. The corruption allegations have been a source of controversy and have led to public discontent. Critics argue that corruption undermines the principles of transparency and accountability in government, eroding public trust. However, it's essential to note that any claims of corruption should be subject to legal scrutiny, and individuals are presumed innocent until proven guilty. Public opinion on this matter may vary, and it's advisable to stay informed through reliable news sources for the latest developments"""
question="what did he faced?"
qa(question=question,context=context)
```

```python
from transformers import AutoTokenizer,TFAutoModelForQuestionAnswering
model_checkpoint="distilbert-base-cased-distilled-squad"
tokenizer=AutoTokenizer.from_pretrained(model_checkpoint)
model=TFAutoModelForQuestionAnswering.from_pretrained(model_checkpoint)
```

```python
inputs=tokenizer(question,context,return_tensors="tf")
output=model(**inputs)
```
