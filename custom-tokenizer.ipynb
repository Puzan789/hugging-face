```python
import tokenizers
```

```python
from tokenizers import (
decoders,
models,
normalizers,
pre_tokenizers,
processors,trainers,
Tokenizer)
```

# Initializing a tokenizer instance

```python
tokenizer=Tokenizer(models.WordPiece(unk_token='[UNK]'))
```

```python
print(dir(tokenizer))
```

```python
print(dir(tokenizer.normalizer))
print(dir(normalizers))
```

# setting the normalization componets converting lowercase to upperce

```python
tokenizer.normalizer=normalizers.Sequence(
[normalizers.Lowercase(),normalizers.NFKD()]
)
```

```python

tokenizer.pre_tokenize=pre_tokenizers.Whitespace()
```

```python
trainer=trainers.WordPieceTrainer(
vocab_size=30_000,
special_tokens=['[UNK]','[PAD]','[CLS]','[SEP]','[MASK]'],
min_frequency=2,
continuing_subword_prefix='##'
)
```

```python
df
```

```python
tokenizer.train_from_iterator(dv_text(),trainer=trainer)
```

```python
tokenizer.get_vocab_size()
```

```python
cls_id=tokenizer.token_to_id('[CLS]')
sep_id=tokenizer.token_to_id('[SEP]')
```

```python
tokenizer.post_processor=processors.TemplateProcessing(
single=f'[CLS]:0 $A:0 [SEP]:0',
pair=f'[CLS]:0 $A:0 [SEP]:0 $B:1 [SEP]:1',
special_tokens=[
    ('[CLS]',cls_id),
    ('[SEP]',sep_id),
    
]
)
```

print(df['text'].shape)
## Final Step is Decoder

```python
tokenizer.decoder=decoders.WordPiece(prefix="##")
```

```python
from transformers import PreTrainedTokenizerFast
full_tokenizer=PreTrainedTokenizerFast(
tokenizer_object=tokenizer,
unk_token='[UNK]',
pad_token='[PAD]',
cls_token='[CLS]',
sep_token='[SEP]',
mask_token='[MASK]',
    
)
```
